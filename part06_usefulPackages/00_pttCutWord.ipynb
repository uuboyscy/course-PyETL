```python
# 引入必要套件
import jieba
import os

# 列出所有文檔
source_file_path = r'./res_gossiping'
file_list = os.listdir(source_file_path)
file_list
```

```python
# 將所有文章存成一個字串
all_article_string = ''
for each_article in file_list:
    article_path = source_file_path + '/' + each_article
    with open(article_path, 'r', encoding = 'utf-8') as f:
        tmp_article_string = f.read().split('---split---')[0].split('\n')[1:]
    for article_line in tmp_article_string:
        all_article_string += article_line + '\n'
print(all_article_string)
```

```python
# 使用預設模式進行分詞
s_list = jieba.cut(all_article_string)
word_count = {}
for i in s_list:
    if i in word_count:
        word_count[i] += 1
    else:
        word_count[i] = 1
word_count
```

```python
'''

加入自定義詞典

'''

mydict_path = r'./mydict.txt'
jieba.load_userdict(mydict_path)

# 使用預設模式進行分詞
s_list = jieba.cut(all_article_string)
word_count = {}
for i in s_list:
    if i in word_count:
        word_count[i] += 1
    else:
        word_count[i] = 1
word_count
```

```python
'''

做進一步篩選
將冗餘字串去除

'''

# 將停用字放在一個list裡面
stopword_path = r'./stopword.txt'
stopword_list = []
with open(stopword_path, 'r', encoding = 'utf-8') as f:
    for each_line in f.readlines():
        stopword_list.append(each_line.replace('\n', ''))
stopword_list

# 只保留長度大於1的詞 len(k) > 1
# 以及不在停用字list裡面的詞 k not in stopword_list
word_list = [(k, word_count[k]) for k in word_count if len(k) > 1 and k not in stopword_list]
word_list
```

```python
# 排序
word_list.sort(key = lambda x : x[1], reverse=True)
word_list
```
