```python
import requests
from bs4 import BeautifulSoup

url = "https://www.104.com.tw/jobs/search/?ro=0&keyword=%E8%B3%87%E6%96%99%E5%88%86%E6%9E%90%E5%B8%AB&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=1&asc=0&page=1&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1%27"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}

res = requests.get(url, headers=headers)

soup = BeautifulSoup(res.text, "html.parser")
```

```python
soup.select('article[class="b-block--top-bord job-list-item b-clearfix js-job-item"]')
```

```python
'''
原程式邏輯
'''

allData = list()
jobArticle_list = soup.select('article[class="b-block--top-bord job-list-item b-clearfix js-job-item"]')
for jobArticle in jobArticle_list:
    data = []
    # print(jobArticle)
    companyContents = jobArticle.select('a')  # <class 'bs4.element.ResultSet'>, list

    companyName = companyContents[1]['title'].split()[0].split("：")[1]
    print("公司名稱:", companyName)
    data.append(companyName)
    data.append("\n")

    jobTitles = jobArticle.select('a em[class="b-txt--highlight"]')
    # print(jobTitles)
    print("應徵職稱: ", end="")
    for jobTitle in jobTitles:
        print(jobTitle.text, end="")
        data.append(jobTitle.text)

    expRequired_list = jobArticle.select('ul[class="b-list-inline b-clearfix job-list-intro b-content"]')
    for expRequired in expRequired_list:
        exp = expRequired.select('li')
        print("\n工作經歷:", exp[1].text, end="\t")
        print("學歷:", exp[2].text)
        data.append("\n")
        data.append(exp[1].text)
        data.append("\n")
        data.append(exp[2].text)

    jobContents = jobArticle.select('p[class="job-list-item__info b-clearfix b-content"]')
    for jobContent in jobContents:
        print(jobContent.text)

        temJC = jobContent.text.split("\n")
        str_a = "_"
        JC = str_a.join(temJC)
        # print(JC)
        data.append("\n")
        data.append(JC)

    salary_list = jobArticle.select('div[class="job-list-tag b-content"]')
    for salary in salary_list:
        sal = salary.select_one('span[class="b-tag--default"]')
        print("\n待遇:", sal.text)
        data.append("\n")
        data.append(sal.text)

    jobContentUrl = 'https:' + companyContents[0]['href']
    print("\n職缺網頁:", jobContentUrl)
    data.append("\n")
    data.append(jobContentUrl)

    companyAddress = companyContents[1]['title'].split()[1]
    print(companyAddress)
    data.append("\n")
    data.append(companyAddress)

    companyUrl = 'https:' + companyContents[1]['href']
    print("公司網頁:", companyUrl)
    data.append("\n")
    data.append(companyUrl)
    
    allData.append(data)
```

```python
'''
會造成每一筆資料長度不一樣，後續資料處理比較複雜
'''
for data in allData:
    print(data.__len__())
```

```python
'''
修改後
'''

allData = list()
jobArticle_list = soup.select('article[class="b-block--top-bord job-list-item b-clearfix js-job-item"]')
for jobArticle in jobArticle_list:
    data = []
    # print(jobArticle)
    companyContents = jobArticle.select('a')  # <class 'bs4.element.ResultSet'>, list

    companyName = companyContents[1]['title'].split()[0].split("：")[1]
    print("公司名稱:", companyName)
    data.append(companyName)
#     data.append("\n")

    jobTitles = jobArticle.select('a em[class="b-txt--highlight"]')
    # print(jobTitles)
    print("應徵職稱: ", end="")
    jobTitleSb = ""
    for jobTitle in jobTitles:
        print(jobTitle.text, end="")
        jobTitleSb += jobTitle.text
    data.append(jobTitleSb)
    
    expSb = ""
    expRequired_list = jobArticle.select('ul[class="b-list-inline b-clearfix job-list-intro b-content"]')
    for expRequired in expRequired_list:
        exp = expRequired.select('li')
        print("\n工作經歷:", exp[1].text, end="\t")
        print("學歷:", exp[2].text)
        expSb += "\n"
        expSb += exp[1].text
        expSb += "\n"
        expSb += exp[2].text
    data.append(expSb)
        
    
    jobContentSb = ""
    jobContents = jobArticle.select('p[class="job-list-item__info b-clearfix b-content"]')
    for jobContent in jobContents:
        print(jobContent.text)

        temJC = jobContent.text.split("\n")
        str_a = "_"
        JC = str_a.join(temJC)
        # print(JC)
        jobContentSb += "\n"
        jobContentSb += JC
    data.append(jobContentSb)

    salarySb = ""
    salary_list = jobArticle.select('div[class="job-list-tag b-content"]')
    for salary in salary_list:
        sal = salary.select_one('span[class="b-tag--default"]')
        print("\n待遇:", sal.text)
        salarySb += "\n"
        salarySb += sal.text
    data.append(salarySb)

    
    jobContentUrl = 'https:' + companyContents[0]['href']
    print("\n職缺網頁:", jobContentUrl)
#     data.append("\n")
    data.append(jobContentUrl)

    companyAddress = companyContents[1]['title'].split()[1]
    print(companyAddress)
    data.append(companyAddress)

    companyUrl = 'https:' + companyContents[1]['href']
    print("公司網頁:", companyUrl)
    data.append(companyUrl)
    
    allData.append(data)
```

```python
"""
可以確保每一筆資料長度都一樣，
除了可保留原始資料外，
還可以確定每個位子都是同樣的欄位
"""

for i in allData:
    print(i.__len__())
```
